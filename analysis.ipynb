```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth

# より広範なスコープを設定
scope = "user-library-read playlist-modify-public user-top-read playlist-read-private"


spotify = spotipy.Spotify(auth_manager=SpotifyOAuth(
    client_id ='74922f440a8a4df6911f8e367923222b', #client ID
    client_secret = '97601c685ab9417da4e3d1904d308f6a', #client secret
    redirect_uri='http://localhost:8888/callback',
    scope=scope
))

try:
    # まずアーティストのトップトラックを取得
    artist_id = '1Xyo4u8uXC1ZmMpatF05PJ'  # The Weeknd
    top_tracks = spotify.artist_top_tracks(artist_id)
    
    if top_tracks and 'tracks' in top_tracks and len(top_tracks['tracks']) > 0:
        seed_track = top_tracks['tracks'][0]['id']
        print(f"Using seed track: {top_tracks['tracks'][0]['name']}")
        
        # market パラメータを追加し、より具体的なパラメータを設定
        recommendations = spotify.recommendations(
            seed_tracks=[seed_track],
            country='US',  # マーケットを指定
            limit=20,
            min_popularity=50,  # 人気度の最小値を設定
            max_popularity=100  # 人気度の最大値を設定
        )
        
        for track in recommendations['tracks']:
            print(f"Track: {track['name']} - Artist: {track['artists'][0]['name']}")
    
    else:
        print("トップトラックの取得に失敗しました")

except spotipy.exceptions.SpotifyException as e:
    print(f"エラーの詳細: {e}")
```
